```python
from explain import *
import yaml
import matplotlib.pyplot as plt
import open3d as o3d
```

```python
n = 501
with open("semantic-kitti.yaml", 'r') as stream:
    semkittiyaml = yaml.safe_load(stream)
```

```python
xai_car = ExplainTarget(target="car")
out_car = xai_car.stats(n_file=n)
```

```python
max_stats_car, predicted_stats_car, confusion_stats_car, fp_instead_stats_car, fn_instead_stats_car, (total_pred_car, total_true_car, total_tp_car) = out_car
```

```python
print(predicted_stats_car.shape)
print(confusion_stats_car.shape)
print(fp_instead_stats_car)
print(fn_instead_stats_car)
print(total_pred_car, total_true_car, total_tp_car)
```

```python
xai_road = ExplainTarget(target="road")
out_road = xai_road.stats(n_file=n)
```

```python
max_stats_road, predicted_stats_road, confusion_stats_road, fp_instead_stats_road, fn_instead_stats_road, (total_pred_road, total_true_road, total_tp_road) = out_road
```

```python
# Accuracy
car_acc = total_tp_car/total_true_car
road_acc = total_tp_road/total_true_road

print("Car accuracy:", car_acc)
print("Road accuracy:", road_acc)
```

# Local

```python
# PCD with lowest avg importance for cars
low_id_car = str(np.argmin(predicted_stats_in_car[:, 0]))
print(low_id_car)

low_file_car = str(low_id_car.rjust(6, "0"))
xyz_file_low_car = f'data/sequences/08/velodyne/{low_file_car}.bin'
points = GradCamPCDPolarNet.read_velodyne_bin(xyz_file_low_car)
xyz_low_car = points[:, :3]
label_file_low_car  = f"data/sequences/08/labels/{low_file_car}.label"

sem_class_to_idx = {v: k for k,v in xai_car.grad_cam.semkittiyaml['labels'].items()}
target_id = sem_class_to_idx["car"]
```

```python
# Grad-CAM

xai_car.grad_cam.load_new_scene("08", low_file_car)
predict_labels_low_car, importances_low_car, heatmap_colors_low_car = xai_car.grad_cam.calculate_grad_cam(target="car")
```

```python
xyz_pol_low_car = GradCamPCDPolarNet.cart2polar(xyz_low_car)
grid_ind_low_car = GradCamPCDPolarNet.get_grid_ind(xyz_pol_low_car)

true_labels_low_car = GradCamPCDPolarNet.load_labels(label_file_low_car)
predict_labels_low_car = GradCamPCDPolarNet.shift_labels(predict_labels_low_car[grid_ind_low_car[:,0], grid_ind_low_car[:,1], grid_ind_low_car[:,2]])

true_colors_low_car = GradCamPCDPolarNet.label2color(true_labels_low_car, colormap=xai_car.grad_cam.semkittiyaml["color_map"])
predict_colors_low_car = GradCamPCDPolarNet.label2color(predict_labels_low_car, colormap=xai_car.grad_cam.semkittiyaml["color_map"])
```

```python
mask = importances_low_car > 0
unique_low_car, counts_low_car = np.unique(predict_labels_low_car[mask], return_counts=True)

percent_low_car = np.asarray([i/counts_low_car.sum() for i in counts_low_car])
unique_low_car_label = np.asarray([xai_car.grad_cam.semkittiyaml['labels'][i] for i in unique_low_car])


inx = np.argsort(percent_low_car)[::-1]

print(dict(zip(unique_low_car_label, percent_low_car)))

plt.figure(figsize=(6, 5))
plt.bar(unique_low_car_label[inx], percent_low_car[inx], color='darkblue')
plt.xticks(rotation=45)
plt.xlabel("Classes")
plt.ylabel("Relative Frequency")
#plt.savefig("fig/not_important_low_car.pdf")
plt.show()
```

```python
# Create Open3D PointCloud object
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(xyz_low_car)
# pcd.colors = o3d.utility.Vector3dVector(true_colors)

# Add colors
pcd.colors = o3d.utility.Vector3dVector(heatmap_colors_low_car)
pcd.colors = o3d.utility.Vector3dVector(true_colors_low_car)
# pcd.colors = o3d.utility.Vector3dVector(predict_colors_low_car)

# Visualize the point cloud
o3d.visualization.draw_geometries([pcd], window_name="GRAD-CAM Point Cloud", width=800, height=600)
```
